```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import time

def browserCreator():
    chrome_options = Options()
    chrome_options.add_argument('--disable-blink-features=AutomationControlled')
    #browser = webdriver.Remote(command_executor="http://127.0.0.1:4444/wd/hub", options=chrome_options)
    browser = webdriver.Chrome(options=chrome_options)
    browser.maximize_window()
    return browser
```

```python
browser = browserCreator()
```

```python
browser.get('https://carrefoursa.com/')
```

```python
sub_categories = browser.find_elements(By.CSS_SELECTOR, 'a.main-menu-dropdown-item-sub-link')

sub_category_urls = []
for href in sub_categories:
    sub_category_urls.append(href.get_attribute('href'))
    print(href.get_attribute('href'))
```

```python
def breadcrumbExtractor(subcategory_url):
    browser.get(subcategory_url)
    breadcrumb_elements = browser.find_elements(By.CSS_SELECTOR, '.breadcrumb li')    
    breadcrumb_list = []
    for category in breadcrumb_elements[1:]:
        breadcrumb_list.append(category.text)
    return breadcrumb_list
```

```python
breadcrumbs = []
broken_breadcrumbs = []
error_count = 0

for category in sub_category_urls:
    try:
        breadcrumb = breadcrumbExtractor(category)
        breadcrumbs.append(breadcrumb)
        error_count = 0
        print(f'SCRAP COMPLETED: {breadcrumb}')
    except:
        print(f'{category} page faced an error!')
        broken_breadcrumbs.append(category)
        time.sleep(3)
        error_count += 1
        if error_count > 5:
            break
```

```python
df = pd.DataFrame(breadcrumbs)
df.to_excel('kategori ağacı.xlsx')
```
